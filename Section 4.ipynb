```python
def my_plot(n):
    for i in range(0,n):
        plt.plot(Games[i], c = 'Black', ls = '--', marker = 's', ms = 10, label = Players[i])
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
my_plot(7)
```

```python
def my_plot(PlayerList):
    c_counter = 0
    for name in PlayerList:
        colors = ['Black', 'Blue', 'Red', 'Green', 'Orange','Magenta', 'Yellow', 'Brown', 'Purple', 'Violet']
        plt.plot(Games[Pdict[name]], c = colors[c_counter], ls = '--', marker = 's', ms = 10, label = name)
        c_counter += 1
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
PlayerList = ["KobeBryant", "LeBronJames", "ChrisBosh"]
my_plot(PlayerList)
```

```python
def my_plot(data,PlayerList=Players):
    c_counter = 0
    for name in PlayerList:
        colors = ['Black', 'Blue', 'Red', 'Green', 'Orange','Magenta', 'Yellow', 'Brown', 'Purple', 'Violet']
        plt.plot(data[Pdict[name]], c = colors[c_counter], ls = '--', marker = 's', ms = 10, label = name)
        c_counter += 1
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```
