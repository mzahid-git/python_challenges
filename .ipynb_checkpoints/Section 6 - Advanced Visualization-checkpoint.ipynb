```python
def sub_cat(data, categorical_variable, by):
    sub_categories = []
    for i in range(0, len(data[categorical_variable].cat.categories)):
        x = data[data[categorical_variable] == data[categorical_variable].cat.categories[i]].by
        sub_categories.append(x)
    return sub_categories
    
```

```python
data = df
categorical_variable = 'Genre'
by=df['Budget']
data[data[categorical_variable] == data[categorical_variable].cat.categories[0]] 

# Note: Cannot reference .Budget in the function because it's being recognized as an object
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style("dark", {"axes.facecolor":"black"}) # white, whitegrid, dark, darkgrid, ticks
f, axes = plt.subplots(4,2,figsize=(15,15))

k1 = sns.kdeplot(data=df, x=df.Budget,y=df.AudienceRatings,shade=True,shade_lowest=False,cmap= 'inferno',ax=axes[0,0])
k1b = sns.kdeplot(data=df, x=df.Budget,y=df.AudienceRatings,cmap= 'cool',ax=axes[0,0])

k2 = sns.kdeplot(data=df, x=df.Budget,y=df.RT_Ratings,shade=True,shade_lowest=False,cmap= 'inferno',ax=axes[0,1])
k2b = sns.kdeplot(data=df, x=df.Budget,y=df.RT_Ratings,cmap= 'cool',ax=axes[0,1])

v1 = sns.boxplot(data=df[df.Genre == 'Drama'],x=df.Year,y=df.RT_Ratings, ax=axes[1,0])

v1 = sns.violinplot(data=df[df.Genre == 'Drama'],x=df.Year,y=df.RT_Ratings,ax=axes[1,1],\
                   palette='cool')

k3 = sns.kdeplot(data=df,x=df.RT_Ratings,y=df.AudienceRatings, shade=True,shade_lowest=False, cmap='Blues_r',ax=axes[2,0])

k3b = sns.kdeplot(data=df,x=df.RT_Ratings,y=df.AudienceRatings, cmap='gist_gray_r',ax=axes[2,0])

k4 = sns.histplot(df.AudienceRatings,ax=axes[2,1])

k5 = sns.histplot(df.RT_Ratings, kde=True, ax=axes[3,0])

axes[3,1].hist(df.AudienceRatings) # non-seaborn plot

k1.set(xlim=(-30,250))
k2.set(xlim=(-30,250))
```
