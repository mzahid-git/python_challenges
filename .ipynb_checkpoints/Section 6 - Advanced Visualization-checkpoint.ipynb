```python
df = df.rename(columns = {'Rotten Tomatoes Ratings %':'RT_Ratings', 'Audience Ratings %':'AudienceRatings',\
                    'Budget (million $)':'Budget', 'Year of release': 'Year'})
```

```python
df.head()
```

```python
df.Year = df.Year.astype('category')
df.Film = df.Film.astype('category')
df.Genre = df.Genre.astype('category')

# Note: category in Pyton is similar to factors in R. We convert obj to category for faster processing and using cat methods.
```

```python
# joint Plots
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.jointplot(data=df, x='RT_Ratings',y='AudienceRatings',)
plt.show()
```

```python
sns.jointplot(data=df, x='RT_Ratings',y='AudienceRatings',kind='hex')
plt.show()
```
